```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

print("All libraries imported successfully! ✅")
```

```python
# Website URL
url = "http://books.toscrape.com/"

# Open the website
response = requests.get(url)

# Check if website opened successfully
print("Status Code:", response.status_code)
```

```python
# Read the HTML of website
soup = BeautifulSoup(response.text, 'html.parser')

# Just to check — print the title of webpage
print(soup.title.text)
```

```python
# Find all books on the page
books = soup.find_all('article', class_='product_pod')

# Check how many books found
print("Number of books found:", len(books))

# Let's see first book data
first_book = books[0]

# Extract title
title = first_book.h3.a['title']

# Extract price
price = first_book.find('p', class_='price_color').text

# Extract rating
rating = first_book.find('p', class_='star-rating')['class'][1]

# Extract availability
availability = first_book.find('p', class_='instock availability').text.strip()

print("Title:", title)
print("Price:", price)
print("Rating:", rating)
print("Availability:", availability)
```

```python
# Empty list to store all books data
all_books = []

# Base URL
base_url = "http://books.toscrape.com/catalogue/page-{}.html"

# Loop through all 50 pages
for page in range(1, 51):

    # Open each page
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all books on this page
    books = soup.find_all('article', class_='product_pod')

    # Extract data from each book
    for book in books:
        title = book.h3.a['title']
        price = book.find('p', class_='price_color').text
        rating = book.find('p', class_='star-rating')['class'][1]
        availability = book.find('p', class_='instock availability').text.strip()

        # Save to list
        all_books.append({
            'Title': title,
            'Price': price,
            'Rating': rating,
            'Availability': availability
        })

    # Show progress
    print(f"Page {page} scraped ✅")

print("\nTotal books scraped:", len(all_books))
```

```python
# Convert list to pandas dataframe
df = pd.DataFrame(all_books)

# Check the data
print("Shape of data:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Save to CSV file
df.to_csv('books.csv', index=False)

print("File saved successfully! ✅")
print("Total books saved:", len(df))
```
