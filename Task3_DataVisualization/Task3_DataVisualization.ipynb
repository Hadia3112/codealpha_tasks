```python
# ============================================
# Task 3: Data Visualization
# Internship: CodeAlpha
# Name: Hadia Ahad
# Date: June 2026
# Description: Visualizing 1000 books dataset
#              from books.toscrape.com
# ============================================

# Import required libraries
import pandas as pd                  # For data manipulation
import matplotlib.pyplot as plt      # For basic charts
import seaborn as sns                # For beautiful charts

# Load dataset
df = pd.read_csv('books.csv')

# Clean Price column
df['Price'] = df['Price'].str.replace('[^0-9.]', '', regex=True).astype(float)

# Convert Rating to numbers
rating_map = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
df['Rating'] = df['Rating'].map(rating_map)

print("Data loaded and cleaned successfully! ✅")
print("Shape:", df.shape)
df.head()
```

```python
# Chart 1 - Rating Distribution
plt.figure(figsize=(10, 5))
sns.countplot(x='Rating', data=df, palette='Blues')

# Add title and labels
plt.title('Distribution of Book Ratings', fontsize=16, fontweight='bold')
plt.xlabel('Rating (1-5)', fontsize=12)
plt.ylabel('Number of Books', fontsize=12)

# Add value labels on top of each bar
for p in plt.gca().patches:
    plt.gca().annotate(str(int(p.get_height())),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.savefig('chart1_rating_distribution.png')
plt.show()
print("Chart 1 saved! ✅")
```

```python
# Chart 2 - Price Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Price'], bins=20, color='skyblue', kde=True)

# Add title and labels
plt.title('Distribution of Book Prices', fontsize=16, fontweight='bold')
plt.xlabel('Price (£)', fontsize=12)
plt.ylabel('Number of Books', fontsize=12)

# Add average price line
plt.axvline(df['Price'].mean(), color='red', linestyle='--', linewidth=2, label=f'Average Price: £{round(df["Price"].mean(), 2)}')
plt.legend(fontsize=11)

plt.tight_layout()
plt.savefig('chart2_price_distribution.png')
plt.show()
print("Chart 2 saved! ✅")
```

```python
# Chart 3 - Average Price per Rating
plt.figure(figsize=(10, 5))
avg_price = df.groupby('Rating')['Price'].mean().reset_index()
sns.barplot(x='Rating', y='Price', data=avg_price, palette='coolwarm')

# Add title and labels
plt.title('Average Price per Rating', fontsize=16, fontweight='bold')
plt.xlabel('Rating (1-5)', fontsize=12)
plt.ylabel('Average Price (£)', fontsize=12)

# Add value labels on top of each bar
for p in plt.gca().patches:
    plt.gca().annotate(f'£{round(p.get_height(), 2)}',
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.savefig('chart3_avg_price_per_rating.png')
plt.show()
print("Chart 3 saved! ✅")
```

```python
# Chart 4 - Top 10 Most Expensive Books
plt.figure(figsize=(12, 6))
top10 = df.nlargest(10, 'Price')
sns.barplot(x='Price', y='Title', data=top10, palette='Reds_r')

# Add title and labels
plt.title('Top 10 Most Expensive Books', fontsize=16, fontweight='bold')
plt.xlabel('Price (£)', fontsize=12)
plt.ylabel('Book Title', fontsize=12)

# Add value labels
for p in plt.gca().patches:
    plt.gca().annotate(f'£{round(p.get_width(), 2)}',
                      (p.get_width(), p.get_y() + p.get_height() / 2.),
                      ha='left', va='center', fontsize=10)

plt.tight_layout()
plt.savefig('chart4_top10_expensive.png')
plt.show()
print("Chart 4 saved! ✅")
```

```python
# Chart 5 - Price vs Rating Boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(x='Rating', y='Price', data=df, palette='Set2')

# Add title and labels
plt.title('Price Distribution per Rating', fontsize=16, fontweight='bold')
plt.xlabel('Rating (1-5)', fontsize=12)
plt.ylabel('Price (£)', fontsize=12)

plt.tight_layout()
plt.savefig('chart5_price_vs_rating.png')
plt.show()
print("Chart 5 saved! ✅")
```

```python
# Chart 6 - Pie Chart of Ratings
plt.figure(figsize=(8, 8))
rating_counts = df['Rating'].value_counts().sort_index()
labels = ['1 Star', '2 Stars', '3 Stars', '4 Stars', '5 Stars']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']
explode = (0.05, 0.05, 0.05, 0.05, 0.05)

plt.pie(rating_counts,
        labels=labels,
        colors=colors,
        explode=explode,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Rating Distribution of Books', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('chart6_pie_ratings.png')
plt.show()
print("Chart 6 saved! ✅")
```

```python
# ============================================
# Summary of Visualizations
# ============================================

print("=" * 50)
print("TASK 3 - DATA VISUALIZATION COMPLETE!")
print("=" * 50)

print("\n📊 Charts Created:")
print("   Chart 1 - Rating Distribution (Bar Chart)")
print("   Chart 2 - Price Distribution (Histogram)")
print("   Chart 3 - Average Price per Rating (Bar Chart)")
print("   Chart 4 - Top 10 Most Expensive Books (Horizontal Bar)")
print("   Chart 5 - Price vs Rating (Boxplot)")
print("   Chart 6 - Rating Distribution (Pie Chart)")

print("\n💡 Key Insights:")
print(f"   Most Common Rating: {df['Rating'].mode()[0]} Stars")
print(f"   Average Book Price: £{round(df['Price'].mean(), 2)}")
print(f"   Most Expensive Book: £{df['Price'].max()}")
print(f"   Cheapest Book: £{df['Price'].min()}")

print("\n✅ All charts saved as PNG files!")
print("=" * 50)
```
